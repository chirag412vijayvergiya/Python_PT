```python
# Single Inheritance
class A:
    def showA(self):
        print("A Show")
class B(A):
    def showB(self):
        print("B Show")

obj1 = A()
obj1.showA()
obj2 = B()
obj2.showA()
obj2.showB()
```

```python
# Multi-Level Inheritance
class A:
    def showA(self):
        print("A Show")
class B(A):
    def showB(self):
        print("B Show")

class C(B):
    def showC(self):
        print("C Show")

obj1 = A()
obj1.showA()
obj2 = B()
obj2.showA()
obj2.showB()
obj3 = C()
obj3.showA()
obj3.showB()
obj3.showC()
```

```python
# Multiple Inheritance
class A:
    def showA(self):
        print("A Show")
class B:
    def showB(self):
        print("B Show")
class C(A, B):
    def showC(self):
        print("C Show")

obj1 = A()
obj1.showA()
obj2 = B()
obj2.showB()
obj3 = C()
obj3.showA()
obj3.showB()
```

```python
# Hierarchical Inheritance
class A:
    def showA(self):
        print("A Show")
class B(A):
    def showB(self):
        print("B Show")
class C(A):
    def showC(self):
        print("C Show")

obj1 = A()
obj1.showA()
obj2 = B()
obj2.showA()
obj2.showB()
obj3 = C()
obj3.showA()
obj3.showC()
```

```python

# Hybrid Inheritance
class A:
    def showA(self):
        print("A Show")
class B(A):
    def showB(self):
        print("B Show")
class C(A):
    def showC(self):
        print("C Show")
class D(B, C):
    def showD(self):
        print("D Show")

obj1 = A()
obj1.showA()
obj2 = B()
obj2.showA()
obj2.showB()
obj3 = C()
obj3.showA()
obj3.showC()
obj4 = D()
obj4.showA()
obj4.showB()
obj4.showC()
obj4.showD()
```

```python
# Method Overriding :- Same name and same parameters
# Main purpose of Method Overriding is to provide specific implementation of the method which is already provided by its superclass
class parent:
    def __init__(self):
        print("Cash, Gold")

    def bike(self):
        print("Splendor+")

class child(parent):
    def bike(self):
        print("Duke 200")

c = child()
c.bike()
```

```python
from abc import ABC, abstractmethod
class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    @abstractmethod
    def m2(self):
        pass

class B(A):
    def m1(self):
        print("m1 method")
        # it is necessary to implement all method which is declared in abstract class
        # if we don't implement all method then it will give error
    def m2(self):
        print("m2 method")

b = B()
b.m1()
b.m2()
```

```python
from abc import ABC, abstractmethod
class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    @abstractmethod
    def m2(self):
        pass

class B(A):
    @abstractmethod
    def m1(self):
        pass
    def m2(self):
        print("m2 method")

class C(B):
    def m1(self):
        print("m1 method")

b = C()
b.m1()
b.m2()
```

```python
class Test1:
    def __init__(self):
        self.a = 10 # public
        self._a = 20 # protected
        self.__a = 30 # private

# protected variable means we can access it in child class but not outside the class
# private variable means we can't access it outside the class but we can access it in class by using _classname__variablename

# Difference between protected and private variable :- 
# protected variable can be accessed in child class but private
# variable can't be accessed in child class

# In java and cpp we can access protected variable in first child class only but in python we can't access protected variable in child class

    def display(self):
        print("a : ", self.a)
        print("_b : ", self._a)
        print("__c : ", self.__a)

class Test2(Test1):
    def show(self):
        print("a : ", self.a)
        print("_b : ", self._a)
        # print("__c : ", self.__a) # it will give error
        print("__c : ", self._Test1__a) # mangling concept :- By using this we can access private variable in child class

t = Test2()
t.display()
t.show()


```

```python
class Employee:
    def __init__(self, name):
        self.name = name
        self.salary = 0

    def work(self):
        print(f"{self.name} is working.")

    def getSalary(self):
        return self.salary

class HRManager(Employee):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary
        self.employees = []

    def work(self):
        print(f"{self.name} is managing HR tasks.")

    def addEmployee(self, employee):
        self.employees.append(employee)
        print(f"Employee {employee.name} added.")

    def showEmployee(self):
        print("Employees under HR Manager:")
        for emp in self.employees:
            print(f"- {emp.name}")

# Example usage
hr_manager = HRManager("Chirag", 5000)
hr_manager.work()
print(f"Salary: {hr_manager.getSalary()}")

emp1 = Employee("Shubham")
emp2 = Employee("Kapil")

hr_manager.addEmployee(emp1)
hr_manager.addEmployee(emp2)
hr_manager.showEmployee()
```

```python
from abc import ABC, abstractmethod
class BankAccount(ABC):
    def __init__ (self):
        self.accno = 0
        self.cname = ""
        self.addr = ""
        self.balance = 0

    @abstractmethod
    def deposit(self):
        pass

    @abstractmethod
    def withdraw(self):
        pass

    def inputCustData(self):
        self.accno = int(input("Enter Account Number : "))
        self.cname = input("Enter Customer Name : ")
        self.addr = input("Enter Address : ")
        self.balance = int(input("Enter Balance : "))

    def displayCustData(self):
        print("Account Number : ", self.accno)
        print("Customer Name : ", self.cname)
        print("Address : ", self.addr)
        print("Balance : ", self.balance)


class SavingAccount(BankAccount):
    def deposit(self, amount):
        self.balance += amount
        print("Amount Deposited Successfully")
        print("Current Balance : ", self.balance)

    def withdraw(self, amount):
        self.balance -= amount
        print("Amount Withdraw Successfully")
        print("Current Balance : ", self.balance)

class currentAccount(BankAccount):
    def deposit(self, amount):
        self.balance += amount
        print("Amount Deposited Successfully")
        print("Current Balance : ", self.balance)

    def withdraw(self, amount):
        self.balance -= amount
        print("Amount Withdraw Successfully")
        print("Current Balance : ", self.balance)


# Menu driven

while True:
    print("1. Saving Account")
    print("2. Current Account")
    print("3. Exit")
    ch = int(input("Enter your choice : "))
    if ch == 1:
        print("_________You are in Saving Account_________")
        obj = SavingAccount()
        obj.inputCustData()
        obj.displayCustData()
        while True:
            print("1. Deposit")
            print("2. Withdraw")
            print("3. Exit")
            ch = int(input("Enter your choice : "))
            if ch == 1:
                amount = int(input("Enter Amount : "))
                obj.deposit(amount)
            elif ch == 2:
                amount = int(input("Enter Amount : "))
                obj.withdraw(amount)
            elif ch == 3:
                break
            else:
                print("Invalid Choice")

    elif ch == 2:
        print("_________You are in Current Account_________")
        obj = currentAccount()
        obj.inputCustData()
        obj.displayCustData()
        while True:
            print("1. Deposit")
            print("2. Withdraw")
            print("3. Exit")
            ch = int(input("Enter your choice : "))
            if ch == 1:
                amount = int(input("Enter Amount : "))
                obj.deposit(amount)
            elif ch == 2:
                amount = int(input("Enter Amount : "))
                obj.withdraw(amount)
            elif ch == 3:
                break
            else:
                print("Invalid Choice")

    elif ch == 3:
        print("Exited Successfully")    
        break

    
```
